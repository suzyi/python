### Create a series of functions that are similar in form, e.g. "f0", "f1", ..., "f10".
August 13, 2018.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m = 10
string = ['f' for i in range(m)]
for k in range(m):
    string[k] = string[k] + str(k)
print string
fun = dict(f0 = 1)
for k in range(m):
    fun[string[k]] = lambda t: k*t
```

```python
T = np.arange(0, 2*np.pi, .1)
plt.figure()
for k in range(m):
    plt.plot(T, fun[string[k]](T))
```
