```python
from scipy.integrate import odeint
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
```

```python
def lorenz(w, t, sigma, rho, beta): 
    x, y, z = w
    return np.array([sigma*(y-x), x*(rho-z)-y, x*y-beta*z])
```

Solve Lorenz system
\begin{eqnarray}
\begin{aligned}
&\frac{dx}{dt}=\sigma(y-x),\\
&\frac{dy}{dt}=x(\rho-z)-y,\\
&\frac{dz}{dt}=xy-\beta z.
\end{aligned}
\end{eqnarray}
with initial condition (the comman choice of parameter is $\sigma=10, \beta=\frac{8}{3}$ and different $\rho$, when $\rho=28$, the system behavior as chaos.)
$$(x_0,y_0,z_0)=(0.0, 1.00, 0.0), (\sigma,\rho,\beta)=(10.0, 28.0, 3.0)$$
and
$$(x_0,y_0,z_0)=(0.0, 1.01, 0.0), (\sigma,\rho,\beta)=(10.0, 28.0, 3.0)$$ respectively.

```python
t = np.arange(0, 30, 0.01) # determine time points
# odeint solver solve two trajectories which has different initial points.
track1 = odeint(lorenz, (0.0, 1.00, 0.0), t, args=(10.0, 28.0, 3.0)) 
track2 = odeint(lorenz, (0.0, 1.01, 0.0), t, args=(10.0, 28.0, 3.0))
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.plot(track1[:,0], track1[:,1], track1[:,2])
ax.plot(track2[:,0], track2[:,1], track2[:,2])
plt.show()
```

```python
# chaos illustration: a slight difference at initial value can cause quite different two curves
plt.figure(figsize=(15, 6))
plt.title('chaos')
plt.subplot(311)
plt.plot(t, track1[:, 0], 'r', t, track2[:, 0], 'b')
plt.ylabel('x(t)')
plt.subplot(312)
plt.plot(t, track1[:, 1], 'r', t, track2[:, 1], 'b')
plt.ylabel('y(t)')
plt.subplot(313)
plt.plot(t, track1[:, 2], 'r', t, track2[:, 2], 'b')
plt.ylabel('z(t)')
plt.xlabel('t')
```
