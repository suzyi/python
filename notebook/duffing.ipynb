```python
# -*- coding: utf-8 -*-
from scipy.integrate import odeint 
import numpy as np 
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
```

```python
def duffing(w, t, alpha, beta, delta): 
    x1, x2= w
    return np.array([x2, -delta*x2-x1*(beta+alpha*np.square(x1))])
```

Solve duffing system
\begin{eqnarray}
\begin{aligned}
&\frac{dx_1}{dt}=x_2,\\
&\frac{dx_2}{dt}=-\delta x_2-x_1(\beta+\alpha x_1^2).
\end{aligned}
\end{eqnarray}
with initial condition
$$(x_1(0),x_2(0))=(-1.2, 0.8), (\alpha,\beta,\delta)=(1, -1.0, 0.5).$$

```python
alpha=1
beta=-1
delta=.5
t = np.arange(0, 30, 0.01) # time span
# odeint solver solve two trajectories which has different initial points.
track1 = odeint(duffing, (-1.2, 0.8), t, args=(1, -1.0, 0.5))
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.plot(track1[:,0], track1[:,1])
plt.show()
```
